```python
# Write a program to create a recursive function to caculate the sum of numbers from 0 to 15
```

```python
def sum_recursive(n):
    if n == 0:
        return 0
    else:
        return n + sum_recursive(n-1)
result = sum_recursive(15)
print('The sum of numbers from 0 to 15 is: ', result)
```

```python
# Import the necessary libraries
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Read the data
```

```python
df = pd.read_csv('company_sales_data.csv')
df
```

```python
# Calculate the total profit for each month 
```

```python
total_profit_per_month = df.groupby('month_number')['total_profit'].sum()

plt.figure(figsize =(10,6))
plt.plot(total_profit_per_month.index, total_profit_per_month.values, marker = 'o', color = 'b')
plt.xlabel('Month Number')
plt.ylabel('Total Profit')
plt.title('Total Profit per Month')
plt.grid(True)
plt.xticks(range(1,13))
plt.show()
```

```python
# Read Bathing soap facewash of all months and display it using the subplot
```

```python

# Calculate the total sales bathing soap and facewash for each month 
total_bathingsoap_per_month = df.groupby('month_number')['bathingsoap'].sum()
total_facewash_per_month = df.groupby('month_number')['facewash'].sum()

#Plot
fig,(ax1, ax2) = plt.subplots(2,1, figsize = (10,12))

#Total sales of bathing soap subplot
ax1.plot(total_bathingsoap_per_month.index, total_bathingsoap_per_month.values, marker = 'o', label = 'BathingSoap ', color = 'y')
ax1.set_xlabel('Month Number')
ax1.set_ylabel('Total Sales')
ax1.set_title('Total Sales of Bathing Soap per Month')
ax1.legend()
ax1.grid(True)
ax1.set_xticks(range(1,13))

# Total sales of facewash subplot
ax2.plot(total_facewash_per_month.index, total_facewash_per_month.values, marker = 'o', label = 'Facewash', color = 'g')
ax2.set_xlabel('Month Number')
ax2.set_ylabel('Total Sales')
ax2.set_title('Total Sales of Bathing Soap and Facewash per Month')
ax2.legend()
ax2.grid(True)
ax2.set_xticks(range(1,13))

plt.tight_layout()
plt.show()
```
